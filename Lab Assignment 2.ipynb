Question 3

```python
import random
from sympy import isprime
random_numbers=[random.randint(100,900) for _ in range(100)]
odd_numbers=[num for num in random_numbers if num%2!=0]
print(odd_numbers)
even_numbers=[num for num in random_numbers if num%2==0]
print(even_numbers)
prime_numbers=[num for num in random_numbers if isprime(num)]
print(prime_numbers)
```

Question 4

```python
A={34,56,78,90}
B={78,45,90,23}
union_scores=A|B
intersection_scores=A&B
symmetric_difference_scores=A^B
is_subset=A<=B
is_superset=B>=A
specific_score=int(input("Enter the score to remove from set A: "))
if specific_score in A:
    A.remove(specific_score)
    print(f"Score {specific_score} removed from set A.")
else:
    print(f"Score {specific_score} is not present in set A.")
print("Union of scores:",union_scores)
print("Intersection of scores:",intersection_scores)
print("Symmetric difference of scores:",symmetric_difference_scores)
print("Is A a subset of B:",is_subset)
print("Is B a superset of A:",is_superset)
```
